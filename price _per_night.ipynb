# Finding the percentile of the price_per_night

```python
# Assume the price of in the data frame is the price per night
max_thresold = df['price'].quantile(0.999)
min_thresold = df['price'].quantile(0.01)

print ("The maximum thresold for the price per night of an apartment/home will be set to: " + str(max_thresold))
print ("The minimum thresold for the price per night of an apartment/home will be set to: " + str(min_thresold))
```

```python
df[df['price']<min_thresold]
```

```python
df[df['price']>max_thresold]
```

```python
#printing the rows that are above maximum thresold and below the minimum thresold

for x in df.index:
    if (df.loc[x, 'price'] > max_thresold) or (df.loc[x, 'price'] < min_thresold):
        print (df.loc[[x]])
        
```

# Creating a new table that does not have outliers

```python
newdf = df[(df['price'] > min_thresold) & (df['price'] < max_thresold)]
newdf.head()
```

```python
newdf.to_string ('clean_AB_NYC_2019.csv')
```
