```python
max_thresold=df['price_per_sqft'].quantile(0.99)
min_thresold=df['price_per_sqft'].quantile(0.001)
```

```python
print ("The maximum thresold is " + str(max_thresold) + " and the minimu thresold is " + str(min_thresold))
```

```python
df[df['price_per_sqft'] < min_thresold]
```

```python
df[df['price_per_sqft']>max_thresold]
```

```python
df.shape
```

```python
newdf= df[(df['price_per_sqft'] < max_thresold) & (df['price_per_sqft'] > min_thresold)]
```

```python
newdf.describe()
```

```python
"""The output below is of dataset that has it outliers removed through the use of pernctile
""" 
print (newdf.to_string())
```
